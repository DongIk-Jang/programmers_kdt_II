```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

  선의 색깔은 노란색(#f2a900) 으로 해야합니다.  
  x-axis label과 y-axis label은 적절한 텍스트로 추가되어야 합니다.  
  이동평균의 기준은 Open Column으로 계산합니다.  
  이외에도 그래프를 알아보기 쉽게 하기 위한 정보(Text, Facet 등)을 추가하는 것도 좋습니다.  
  💡 이동평균(Moving Average)법은 시계열 데이터를 표현하는 데에 많이 사용하는 방법 중 하나입니다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
BTC_df = pd.read_csv("./BitCoin.csv")
BTC_df = BTC_df.sort_values('Date')
BTC_df
```

```python
BTC_df.isnull().sum()
```

```python
BTC_df = BTC_df[BTC_df['Date'] >= '2016-05-26']
BTC_df = BTC_df[BTC_df['Date'] < '2017-07-01']

BTC_df
```

```python
BTC_df.head(10)
```

```python
BTC_ma = BTC_df['Open'].rolling(5).mean()
BTC_ma = BTC_ma[6:]
BTC_df = BTC_df[6:]
BTC_ma.head(10)
```

```python
BTC_ma.count()
```

```python
plt.figure(figsize=(15, 10))
plt.xlabel("Every 30 Days")
plt.ylabel('Price(Dollars)')
plt.title('BitCoin 5-MA Graph')
plt.xticks([i for i in range(0, 391, 30)])


plt.plot(BTC_df['Date'], BTC_ma, color='#f2a900')
plt.show()
```

- 2016.6 ~ 2017.6 기간에 해당하는 비트코인의 5-MA 그래프와 이더리움의 5-MA 그래프를 동시에 그려주세요.

  선의 색깔은 비트코인 : 노란색(#f2a900), 이더리움 : 검은색(#3c3c3d) 으로 해야합니다.  
  x-axis label과 y-axis label은 적절한 텍스트로 추가되어야 합니다.  
  이동평균의 기준은 Open Column으로 계산합니다.  
  이외에도 그래프를 알아보기 쉽게 하기 위한 정보(Text, Facet 등)을 추가하는 것도 좋습니다.

```python
eth_df = pd.read_csv('./ETH_day.csv')
eth_df.dtypes
```

```python
eth_df.isnull().sum()
```

```python
eth_df = eth_df.sort_values(by=['Date'])
eth_df
```

```python
eth_df = eth_df[eth_df['Date']>='2016-05-26']
eth_df = eth_df[eth_df['Date']<'2017-07-01']
eth_df = eth_df.sort_values(by=['Date'])
eth_df.head(10)
```

```python
eth_ma = eth_df['Open'].rolling(5).mean()
eth_ma = eth_ma[6:]
eth_df = eth_df[6:]
eth_ma.head(10)
```

```python
plt.figure(figsize=(15, 10))
plt.xlabel("Every 30 Days")
plt.ylabel('Dollars')
plt.title('BitCoin 5-MA Graph')
plt.xticks([i for i in range(0, 391, 30)])
plt.grid()

plt.plot(eth_df['Date'], eth_ma, color='#3c3c3d')
plt.plot(BTC_df['Date'], BTC_ma, color='#f2a900')

plt.legend(['ETH 5-MA', 'BTC 5-MA'])
plt.show()
```
