## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- matplotlib 는 기본 모듈이 아니고 numpy와 pandas처럼 추가적인 설치가 필요하다.
- `%matplotlib inline`을 통해서 활성화

cd ) 라이브러리 vs 프레임워크  
라이브러리는 개발자들이 만들어만 놓았고 우리가 라이브러리 내부의 코드들을 조합해서 결과물을 만들어가야 하고,  
프레임워크는 틀이 이미 만들어져있고 우리는 그 틀 안에 내용물을 채워가는 방식으로 결과물을 만들어낸다.  
  
라이브러리로는 numpy, pandas, matplotlib 등이 있고, 프레임워크로는 django, flask 등이 있다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study

### Case Study with Arguments

- `.plot()` 함수는 꺽은선 그래프를 그려주는 함수

```python
plt.plot([1, 2, 3, 4, 5]) # 실제 plotting을 하는 함수
# y = x + 1 그래프
# [1, 2, 3, 4, 5] 부분이 y값에 해당하고 따로 x에 대해 명시하지 않았기 때문에 각 인덱스 값이 x값으로 들어갔다.
plt.show() # plt를 확인하는 명령
```

```python
plt.plot([2, 4, 2, 4, 2])
plt.show()
```

### Figsize : Figure(도면)의 크기 조절

```python
plt.figure(figsize=(3, 3)) # plotting을 할 도면을 선언

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y=x를 그려보면:

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
# numpy.array를 이용해서 함수 그래프 그리기
# y = x^2

x = np.array([0, 1, 2, 3, 4]) # 정의역
y = np.array([0, 1, 4, 9, 16]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) 파이썬에서는 c에 정수가 들어갔었는데 우리는 c를 매우 작은 값으로 쪼개서 연속적인 그래프를 그리고싶음

x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.plot(x, x**2)

plt.show()
```

```python
# x, y 축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.plot(x, x**2)

plt.show()
```

```python
# x, y축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 25, 3)]) # y축의 눈금 설정, 0, 3, 6, ...

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 25, 3)]) # y축의 눈금 설정, 0, 3, 6, ...

plt.title("y = x^2 graph")

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프에 legend(범례) 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 25, 3)]) # y축의 눈금 설정, 0, 3, 6, ...
plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trend") # 그래프의 선에 label을 지정해주기
# legent(범례) 지정은 꼭 plot 그린 이후에 작성
plt.legend()

plt.show()
```

## III. Matplotlib Case Study

### 꺾은 선 그래프 (Plot)

- `.plot()`

```python
x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 0~19 사이의 난수를 20번 생성
# 그냥 random.randint(0, 20) 이라고만 하면 0~20 사이의 난수 1개 생성
# 같은 함수인데도 세 번째 parameter 여부에 따라 범위에 차이가 있는 듯

plt.plot(x, y)
plt.show()
```

```python
# Extra : y축을 20까지 보이게 하고, y축을 5 단위로 보이게 하고싶다면?

x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20)

plt.axis([0, 20, 0, 20])
plt.xticks([i for i in range(0, 20)])
plt.yticks([i for i in range(0, 21, 5)])

plt.plot(x, y)
plt.show()
```

### 산점도 (Scatter Plot)

- `.scatter()`

위의 꺾은 선 그래프와 사용 용도 면에서 어떻게 다를까?  
- 꺾은선 그래프 (Plot)  
  시계열 데이터에서 많이 사용된다. x축이 시간인 경우.  
  이전 시간에 비해 다음 값이 얼마나 달라졌나를 확인하는 목적.  
- 산점도 (Scatter Plot)  
  x, y가 완전히 별개의 변수인 경우 많이 사용.  
  산점도를 그려봤을 때 어떤 상관관계가 보이는 지 확인하는 용도.

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

- `.boxplot()`
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot(y)
plt.show()
```

```python
# Extra : Plot의 title을 "Box plot of x, y"

plt.boxplot((x, y)) # 두 개의 인자를 컨테이너에 담아서 인자로 넣어서 x, y의 분포를 비교해볼 수 있음
plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)

- `.bar()`
- 범주형 데이터의 값과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)
plt.xticks(np.arange(0, 20, 1))

plt.show()

# Extra : xticks를 올바르게 처리해봅시다.
```

### cf) 히스토그램 (Histogram)

- `.hist()`
- 막대 그래프와의 공통점 : 도수분포를 직사각형의 막대 형태로 나타낸다.
- 차이점 : "계급"으로 나타내는 것이 특징. 0, 1, 2가 아니라 0~2까지의 "범주형" 데이터로 구성 후 그림을 그림.
- histogram은 범주가 연속성을 갖는 경우가 많아서 직사각형이 연속적으로 연결되어있음.

```python
plt.hist(y, bins=np.arange(0, 22, 2))
plt.xticks(np.arange(0, 22, 2))

plt.show()

# Extra : xticks를 올바르게 고쳐봅시다.
```

### 원형 그래프 (Pie Chart)

- `.pie()`
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인이 용이

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four']) # pit chart에서는 특히 라벨을 붙여줄 필요가 있음
plt.show()
```

### Seaborn Import 하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)

- `sns.kdeplot()`
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림

```python
# in Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade=True)

plt.show()
```

### 카운트그림 (Count Plot)

- `sns.countplot()`
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과

```python
vote_df = pd.DataFrame({"name":['Andy', 'Bob', 'Cat'], "vote":[True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])

plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df['vote'])

plt.show()
```

### 캣그림 (Cat Plot)

- `sns.catplot()`
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
sns.catplot(x='WHO Region', y='Confirmed', data=covid)
plt.show()
```

```python
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin') # kind는 어떤 그래프 형태로 보여줄 지 정해줌
s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립그림 (Strip Plot)

- `sns.stripplot()`
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프

```python
sns.stripplot(x='WHO Region', y='Recovered', data=covid)

plt.show()
```

### cf) Swarmplot

- stripplot에서는 뭉쳐있는 부분에 얼마나 많은 데이터가 뭉쳐있는 지 확인하기 어렵다
- swarmplot에서는 뭉쳐있는 부분을 가로로 펼쳐서 표시해줌으로써 얼마나 많은 데이터가 뭉쳐있는 지 확인하기 용이하다.

```python
s = sns.swarmplot(x='WHO Region', y='Recovered', data=covid)

plt.show()
```

### 히트맵 (Heatmap)

- `sns.heatmap()`
- 데이터의 행렬을 색상으로 표현해주는 그래프
- 가장 많이 사용하는 예시는 상관계수행렬(correlation matrix)이다.

```python
# 히트맵 예제

covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()

# 밝은 곳에서 큰 상관관계가 있고 어두운 곳은 상관관계가 약하다.
```
